# II- Egalisation de l'histogramme

### Import libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Lecture d'image

```python

# charger l'image dans la matrice img
path = 'source/eight.tif'
img = cv.imread(path, 0)
```

## II.1- Egalisation Globale

```python
img1 = cv.equalizeHist(img)
```

## II.2- Egalisation Locale

```python
clahe = cv.createCLAHE()
img2 = clahe.apply(img)
```

## Affichage

```python
# affichage des images
cv.imshow('Originale', img)
cv.imshow('Egal Globale', img1)
cv.imshow('Egal Locale', img2)

# calcul et affichage de l'histogramme
fig, ax = plt.subplots(1, 3, figsize=(20,10))

ax[0].hist(img.ravel(), 256, [0, 256])
ax[0].set_title(label="Originale")

ax[1].hist(img1.ravel(), 256, [0, 256])
ax[1].set_title(label="Apres Egalisation Globale")

ax[2].hist(img2.ravel(), 256, [0, 256])
ax[2].set_title(label="Apres Egalisation Locale")

plt.show()

# attendre la fermeture des fen√®tres
cv.waitKey(delay=0)
cv.destroyAllWindows()
```
